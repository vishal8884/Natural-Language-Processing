```python
from nltk.stem import WordNetLemmatizer
import nltk
# Download the wordnet corpus
nltk.download('wordnet')

words = [
    "hello", "sit", "sitting", "bye", "programs", "tata", "will be back",
    "gamer", "gaming", "coder", "coding", "programming", "program", "programmer",
    "runnable", "running", "ran","failure","goes","swapping","girls","men","woman"

]

lemmatizer = WordNetLemmatizer()

for word in words:
    pos = 'v' if word.endswith('ing') or word in ['ran', 'sit'] else 'n'
    print(f"{word} ---> {lemmatizer.lemmatize(word, pos=pos)}")
```
