# Import the tokenizer and lemzatizer and stopwards

```python
from nltk.tokenize import sent_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()
```

# Stopwards Demo

```python
stop_words = set(stopwords.words('english'))  #This gives all the words which is not having more meaning
print(stop_words)
```

Tokensize the speech


```python
tokenizedSpeech = sent_tokenize(speech)
type(tokenizedSpeech)
```

```python

for sentence in tokenizedSpeech:
  words = nltk.word_tokenize(sentence) #Born, in , ranchi ...
  updatedSentense = ''
  for word in words:
    if word not in stop_words:
        lemmatizedWords = lemmatizer.lemmatize(word);
        updatedSentense = updatedSentense + ' ' + lemmatizedWords
  print(updatedSentense)

```
